```python
#Escrever um programa que descobre se um numero é par ou impar?

#Entrada
numero = float (input("digite um numero: "))

#Processamento
if numero / 2 == 0:
  print (f"o número {numero}, é um número par!!!")

else:
  print (f"o número {numero}, é um número impar!!!")
```

```python
#Escrever um programa que verifica se um número é maior, menor ou igual
#ao outro.

#Entrada
a = float(input("digite um numero: "))
b = float(input("digite um numero: "))

#Processamento

if a > b :
  print(f"{a} é maior que {b}")
elif a == b :
  print(f" {a} é igual a {b}")
else:
  print(f" {a} é menor que {b}")

```

```python
#Escrever um programa que recebe 04 notas do aluno calcula a média e mostra se o
#aluno foi aprovado
#Se a nota for >7 aprovado!!!
#Se a nota for >=5 você esta de recuperação!!
#se a nota for <5 você for reprovado!

#Entrada
print(f"-----------------------Digite suas notas aqui!!!----------------")
print(f"--------------------------Notas de 1 a 10!!!--------------------------")
nota1 = float (input(" digite sua nota do 1 bimestre: "))
nota2 = float (input(" digite sua nota do 2 bimestre: "))
nota3 = float (input(" digite sua nota do 3 bimestre: "))
nota4 = float (input(" digite sua nota do 4 bimestre: "))

#Processamento

resultado = (nota1 + nota2 + nota3 +nota4) / 4

#Saída

if resultado >7:
  print(f"A média das usas notas foi {resultado}, você foi aprovado!!!")
elif resultado >= 5:
  print(f"A média das usas notas foi {resultado}, você terá que fazer recuperação!!")
else:
  print(f"A média das usas notas foi {resultado}, você foi reprovado!")
```

```python
# Peça o valor de uma compra: ≥ 100: 10% de desconto, ≥ 50 e < 100: 5% de desconto , < 50: sem desconto
# entrada
compra = float (input("digite o valor da compra:  "))

# processamento
if compra >= 100:
  valor_final = compra * 0.90
  print(f"O valor final da compra com 10% de desconto é: R$ {valor_final:.2f}")

elif compra >= 50:
  valor_final = compra * 0.95
  print(f"O valor final da compra com 5% de desconto é: R$ {valor_final:.2f}")

else:
  valor_final = compra
  print(f"O valor final da compra é: R$ {valor_final:.2f}, não possui desconto")
```
