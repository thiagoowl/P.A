```python
#Operadores aritméticos
# Fazer um programa que o usuário digite 02 numeros.
# Saída mostrar operações artiméticas.

# Entrada
a = int (input("digite um numero: "))
b = int (input("digite um numero: "))

# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a/b
div_int = a //b
resto = a % b
potencia = a ** b
# Saída
print("")
print("-------------------------OPERADORES ARITMÈTICOS-----------------------|")
print("                                                                      |")
print("----------------------------------SOMA---------------------------------|")
print("                                                                       |")
print(f"o resultado da SOMA de {a} mais {b} é igual a: {soma}")
print("")
print("-----------------------------------------------------------------------|")
print("-------------------------OPERADORES ARITMÈTICOS-----------------------|")
print("                                                                      |")
print("----------------------------------SUBTRAÇÃO---------------------------------|")
print("                                                                       |")
print(f"o resultado da SUBTRACAO de {a} menos {b} é igual a: {subtracao}")
print("")
print("-----------------------------------------------------------------------|")
print("-------------------------OPERADORES ARITMÈTICOS-----------------------|")
print("                                                                      |")
print("--------------------------------MULTIPLICAÇÃO---------------------------------|")
print("                                                                       |")
print(f"o resultado da MILTIPLICACAO de {a} vezes {b} é igual a: {multiplicacao}")
print("")
print("-----------------------------------------------------------------------|")
print("                                                                      |")
print("----------------------------------DIVISÃO---------------------------------|")
print("                                                                       |")
print(f"o resultado da DIVISAO de {a} dividido {b} é igual a: {divisao}")
print("")
print("-----------------------------------------------------------------------|")
print("-----------------------------------------------------------------------|")
print("                                                                      |")
print("----------------------------------DIVISÃO---------------------------------|")
print("                                                                       |")
print(f"o resultado da DIVISAO INTERIA de {a} dividido {b} é igual a: {div_int}")
print("")
print("-----------------------------------------------------------------------| ")
print("-----------------------------------------------------------------------|")
print("                                                                      |")
print("----------------------------------DIVISÃO---------------------------------| ")
print("                                                                       |")
print(f"o RESTO da DIVISAO de {a} dividido {b} é igual a: {resto}")
print("")
print("-----------------------------------------------------------------------|")
print("                                                                      |")
print("----------------------------------POTENCIAÇÂO---------------------------------| ")
print("                                                                       |")
print(f"o resultado da POTENCIAÇÃO de {a} elevado {b} é igual a: {potencia}")
print("")
print("-----------------------------------------------------------------------|")

```

```python
# Fazer Programa Que Converte Horas em Minutos e Horas em Segundos

# Entrada
horas = int (input("digite um numero: "))

# Processamento
Minutos = horas * 60
Segundos = horas * 3600

# Saida
print("")
print("-------------------------HORAS EM MINUTOS-----------------------|")
print("                                                                      |")
print("----------------------------------------------------------------------|")
print("                                                                       |")
print(f"o resultado de Minutos é igual a: {Minutos}")
print("")
print("-----------------------------------------------------------------------|")
print("")
print("-------------------------HORAS EM SEGUNDOS-----------------------|")
print("                                                                      |")
print("----------------------------------------------------------------------|")
print("                                                                       |")
print(f"o resultado de Segundos é igual a: {Segundos}")
print("")
print("-----------------------------------------------------------------------|")
```
