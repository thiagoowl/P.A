```python
 # Escrever um programa que verifica se um numero é par ou impar?

 #Entrada
numero = int (input("digite um numero: "))

#Processamento
if numero % 2 == 0:
  print (f"o número {numero}, é um número par.")

else:
   print (f"o número {numero}, é um número impar.")
```

```python
# Escrever um programa que o usuario digita o nome e as suas 04 notas referantes aos bimestres, calcula a média e mostra
# aprovado, recuperação ou reprovado.
# nota maior ou igual a 7 aprovado
# nota maior ou igual 5 e menor que 7 recuperação
# nota menor que 5 reprovado


# Entrada
nome = str (input("Digite seu nome: "))

nota1 = float (input("digite a sua nota do 1  bimestre: "))
if nota1 > 10:
  print("Digite uma nota valida (notas de 0 a 10)")
nota2 = float (input("digite a sua nota do 2  bimestre: "))
if  nota2 > 10:
  print("Digite uma nota valida (notas de 0 a 10)")
nota3 = float (input("digite a sua nota do 3  bimestre: "))
if  nota3 > 10:
  print("Digite uma nota valida (notas de 0 a 10)")
nota4 = float (input("digite a sua nota do 4  bimestre: "))
if  nota4 > 10:
  print("Digite uma nota valida (notas de 0 a 10)")

média = (nota1 + nota2 + nota3 + nota4) /4

# Saida
print (f"-------------------------------{nome}---------------------------------")

if média >= 7:
 print (f"A média das suas notas foi {média}, você foi aprovado!")
elif média >= 5:
     print (f"A média das suas notas foi {média}, você tera que fazer recuperação!")
else:
     print (f"A média das suas notas foi {média}, você foi reprovado!")
```
