```python
# Escrever um programa que recebe 04 notas do aluno calcula a média e mostra se o aluno foi aprovado.

#Entrada
print ("digite suas notas AQUI, valores de 1 a 10!")
nota1 = float (input("Digite a nota do 1 bimestre: "))
nota2 = float (input("Digite a nota do 2 bimestre: "))
nota3 = float (input("Digite a nota do 3 bimestre: "))
nota4 = float (input("Digite a nota do 4 bimestre: "))

#Processamento

resultado = (nota1 + nota2 + nota3 + nota4) / 4

if resultado >= 7:

     print (f"A média das suas notas foi {resultado}, você foi aprovado!")
elif resultado >= 5:
     print (f"A média das suas notas foi {resultado}, você tera que fazer recuperação!")
else:
     print (f"A média das suas notas foi {resultado}, você foi reprovado!")
```
